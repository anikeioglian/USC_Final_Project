```python
import zipfile

with zipfile.ZipFile("/kaggle/input/dogs-vs-cats/train.zip","r") as zip_ref:
    zip_ref.extractall("dogs-vs-cats")
    
with zipfile.ZipFile("/kaggle/input/dogs-vs-cats/test1.zip","r") as zip_ref:
    zip_ref.extractall("dogs-vs-cats")
```

```python
import os
train_file = os.listdir('/kaggle/working/dogs-vs-cats/train')
test_file = os.listdir('/kaggle/working/dogs-vs-cats/test1')
```

```python
animal_type = []
# We only have cats and dogs in this data set, so let's organize these images
# Each image in the zip file has the format 'cat.#.jpg' or 'dog.#.jpg'
# Append each category of animal into the list by their names
# Test file does not have the same format, only make data fame for train file for now.
for image in train_file:
    if image.split('.')[0] == 'dog':
        animal_type.append('dog')
    else:
        animal_type.append('cat')
# Now, let's turn our list into a training dataframe to begin this process!        
train_df = pd.DataFrame({'images': image, 'label': animal_type})
print(train_df.head())
```

```python
# Let's explore our data
shape_df = train_df.shape
print(shape_df)
```

```python
# Check descriptive statistics and number of cats vs. dogs
print(train_df.describe())
print(train_df.label.value_counts())
```
